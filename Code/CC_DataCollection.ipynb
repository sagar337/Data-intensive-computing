```python
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
import re
import nltk
nltk.download('stopwords')
import io 
from nltk.corpus import stopwords 
```

```python
stop_words =set(stopwords.words('english')) 
#file = open('commonCrawlData.txt', encoding="utf8")
with open('commonCrawlData.txt','r',encoding="utf8") as f:
        file= open('CCResult.txt','a')
        for line in f:
            singles = []
            
            stemmer = PorterStemmer() 
            for plural in line.split():
                singles.append(stemmer.stem(plural))
                #file.write(' '.join(singles))
            #print(' '.join(singles))
            res = ' '.join(singles)
            cleanText = re.sub(r'[^\w\s]','',res)
            resulttext = re.sub(r'[0-9]', '', cleanText)
            print("cleantext")
            #print(cleanText)
            words = resulttext.split()
            for r in words: 
                if not r in stop_words:
                    print(r)
                    file.write(" "+r) 
            
        file.close()
                
            
```
